### Adding Grade using 
    -List Comprehension
    -Function

```python
df["Grade"] = [
    "A" if a >= 90
      else "B" if a >= 80
        else "C" if a >= 70
          else "D" if a >=60
            else "F"
              for a in df["Percentage"]]
df
```

    -USING FUNCTION

```python
def calculateGrade(percentage):
    if percentage >= 90:
        return "A+"
    elif percentage >= 80:
      return "A"
    elif percentage >= 70:
       return "B"
    elif percentage >= 60:
       return "C"
    elif percentage >= 50:
       return "D"
    else:
       return "F"
    
df["Grade"] = df["Percentage"].apply(calculateGrade) 
df
```
