```python
# data[row:col] ---> str end
# iloc => integer loc
# rows ko slice or column ko slice ek sath kre to iloc use krte
data.iloc[2:5, 1:4]
# column index starts from 0 too
```

    -.loc is label-based: You select data using the explicit names of the rows or columns.

    -.iloc is integer-position-based: You select data using the implicit 0-indexed position (like a standard Python list), completely ignoring the actual names.
