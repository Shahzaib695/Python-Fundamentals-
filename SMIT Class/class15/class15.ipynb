```python
class Car:
    def __init__(self,name,brand,wheel,engine,type,interior,year,color):
        self.name = name
        self.brand = brand
        self.wheel = wheel
        self.engine = engine
        self.type = type
        self.interior = interior
        # self.model = model 
        self.year = year
        self.color = color
    def idle(self):
            print(f"{self.brand} {self.name} is idle");
    def start(self):
            print(f"{self.brand} {self.name} is stated");
    def stop(self):
            print(f"{self.brand} {self.name} is stopped");
    def wheels(self):
            print(f"{self.brand} {self.name} have {self.wheel} wheels");
    def colors(self):
            print(f"{self.brand} {self.name} is of {self.color} color");
    def modife_color(self,new_color):
           self.color = new_color
           print(f"The color is changed your new color is {self.color}")
```

```python
Car1 = Car("Vitz","Toyota",4,"1000cc","Petrol","Biege",2025,"white");
Car1.idle()
Car1.colors()
Car1.wheels()
Car1.modife_color("Black")
```
