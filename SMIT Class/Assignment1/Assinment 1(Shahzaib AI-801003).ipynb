# 🔹 Q14 (Marks + Total) User se 3 subjects ke marks lo aur total print karo using f-string.

```python
# type your code here
sub1 = int(input("Enter Marks Of Subject 1: "));
sub2 = int(input("Enter Marks Of Subject 2: "));
sub3 = int(input("Enter Marks Of Subject 3: "));
total = sub1 + sub2 + sub3
print(f"The total marks of all three subjects are: {total}")
```
