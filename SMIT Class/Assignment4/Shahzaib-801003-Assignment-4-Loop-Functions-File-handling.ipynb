**Q12: Check karein ke number Prime hai ya nahi.**

```python
number = int(input("Koi number enter karein: "))

if number > 1:
    prime = True

    for i in range(2, number):
        if number % i == 0:
            prime = False
            break

    if prime:
        print("Number Prime hai")
    else:
        print("Number Prime nahi hai")
else:
    print("Number Prime nahi hai")
```

**Q23: Arbitrary arguments (`*args`) ka use karke function banayein jo saare numbers ko multiply kare.**

```python
def multiply_numbers(*args):
    result = 1

    for number in args:
        result *= number

    return result


print(f"Multiplication of (2, 3, 4) is: {multiply_numbers(2, 3, 4)}")

print(f"Multiplication of (5, 10) is: {multiply_numbers(5, 10)}")
```

**Q24: Keyword arguments (`**kwargs`) ka use karke student ki info print karein (Name, Age).**

#### **Q29: Recursive function banayein jo factorial calculate kare.**

```python
def factorial(number):
    if number == 1 or number == 0:
        return 1
    else:
        return number * factorial(number - 1)


num = int(input("Enter a number: "))

print(f"The factorial of {num} is: {factorial(num)}")
```

#### **Q30: Function banayein jo user se unlimited numbers le (*args) aur sirf even numbers ki list return kare.**

```python
def even_numbers(*args):
    even_list = []

    for number in args:
        if number % 2 == 0:
            even_list.append(number)

    return even_list


print(f"Even numbers from (1, 2, 3, 4, 5, 6, 7, 8) are: {even_numbers(1, 2, 3, 4, 5, 6, 7, 8)}")
```
