### Indexing And Slicing

```python
# Indexing = accessing of value
# slicing = aik copy bana deti hai
```

```python
# np.arange(64) creates a 1D array with values from 0 to 63
# reshape(8, 8) converts that 1D array into an 8x8 2D array
data = np.arange(64).reshape(8, 8)
data
```

```python
data[5,3]
```

```python
data[5,-1]
# or
data[5,7]
```

```python
# [row, col]
# [st:end,st:end]
```

```python
# Select row 4, and columns 2 through 3 (slice 2:4 excludes index 4)
data[4, 2:4]
```

```python
data[4:6, 3:4]
```

```python
data2 = np.arange(64).reshape(4,4,4)
data2
```

```python
data2[2][2,1:]
```
