```python
import numpy as np
```

```python
np.array([1,2,3,4,5,6])
```

```python
# if i will not give it in list there will be an error because array function takes list as an argument and convert it into array
arr = np.array([1,2,3,4,5,6])
arr.size
```

```python
arr.shape # it will give us the shape of the array in tuple form (cols,rows)
type(arr) # it will give us the type of the array
arr.dtype # it will give us the data type of the array
arr.ndim # it will give us the number of dimensions of the array
arr.nbytes # it will give us the total number of bytes consumed by the array
arr.itemsize # it will give us the number of bytes consumed by each element of the array
```

```python
arr2 = np.array([[1,2,3],[4,5,6]])
arr2.shape
```

```python
arr2.ndim
```
