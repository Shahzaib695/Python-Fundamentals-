```python
class parent:
    def __init__(self):
        self.name = "parent"

    def get_name(self):
        return self.name

class child(parent):
    def __init__(self):
        super().__init__()
        self.name = "child"
```

```python
# polymorphism
class Parent:
    def work(self):
        return "Parent is working"

class ChildA(Parent):
    def work(self):
        return "ChildA is playing"

class ChildB(Parent):
    def work(self):
        return "ChildB is studying"

def do_work(person: Parent):
    print(person.work())

people = [Parent(), ChildA(), ChildB()]

for person in people:
    do_work(person)
```

```python
# inheritance
class Parent:
    def __init__(self, name):
        self.name = name

    def greet(self):
        return f"Hello, I am {self.name}."

class Child(Parent):
    def __init__(self, name, age):
        super().__init__(name)
        self.age = age

    def greet(self):
        return f"Hello, I am {self.name} and I am {self.age} years old."

parent = Parent("Ali")
child = Child("Sara", 12)

print(parent.greet())
print(child.greet())

# explanation:
# - Child inherits from Parent and reuses Parent's __init__ method.
# - super().__init__(name) calls the Parent constructor to set the name.
# - Child adds its own age attribute and overrides greet() with new behavior.
# - This example shows inheritance, code reuse, and method overriding.
```

### Abstraction :
--Python does not support abstraction we can use it using a module named as ABC for abstraction. ABC means abstract base classes

```python
from abc import ABC, abstractmethod

class Computer(ABC):
    @abstractmethod
    def process(self):
        pass

    @abstractmethod
    def greet(self):
        pass

class Laptop(Computer):
    def process(self):
        return "Laptop is processing data"

    def greet(self):
        return "Hello from Laptop"

class Phone(Computer):
    def process(self):
        return "Phone is processing a call"

    def greet(self):
        return "Hello from Phone"

for device in (Laptop(), Phone()):
    print(device.greet())
    print(device.process())
```

### Abstraction example explanation
- `Computer` is an abstract base class created with `ABC`.
- `process()` and `greet()` are abstract methods, so subclasses must implement them.
- `Laptop` and `Phone` inherit from `Computer` and provide their own concrete behavior.
- This shows abstraction by defining an interface in the parent and hiding implementation details in the children.

```python
from abc import ABC, abstractmethod

class Car(ABC):
    @abstractmethod
    def start_engine(self):
        pass

    @abstractmethod
    def stop_engine(self):
        pass

class Sedan(Car):
    def start_engine(self):
        return "Sedan engine started"

    def stop_engine(self):
        return "Sedan engine stopped"

class Truck(Car):
    def start_engine(self):
        return "Truck engine started"

    def stop_engine(self):
        return "Truck engine stopped"

for vehicle in (Sedan(), Truck()):
    print(vehicle.start_engine())
    print(vehicle.stop_engine())

# explanation:
# - Car is an abstract base class with abstract methods start_engine() and stop_engine().
# - Sedan and Truck inherit from Car and implement both abstract methods.
# - You cannot create a Car directly, only concrete subclasses that define the methods.
# - This demonstrates abstraction: the parent defines the interface and the children provide the implementation.
```

```python
from abc import ABC, abstractmethod

class Animal(ABC):
    @abstractmethod
    def sound(self):
        pass

class Dog(Animal):
    def sound(self):
        print("Woof")

# create and run
pet = Dog()
pet.sound()
```

### Abstraction example (Animal)
- `Animal` is an abstract base class.
- `sound()` is abstract and must be implemented by child classes.
- `Dog` inherits from `Animal` and implements `sound()`.
- This shows abstraction: the parent defines the interface, the subclass defines the behavior.
