```python
# Student id card
name = input("Enter Your Name: ").title();
father_name = input("Enter Your Father Name: ").title();
rollNo = int(input("Enter Your Roll Number"));
course = input("Enter Your Course Name: ").title();
institue = input("Enter Your Institue Name: ").title();
branch = input("Enter Your Branch Name: ").title();
print(f'''
      
            STUDENT ID CARD
      ============================
      Name => {name}.
      Father name => {father_name}.
      Roll number => {rollNo}.
      Course => {course}.
      Institute => {institue}.
      Branch => {branch}.''')
```
