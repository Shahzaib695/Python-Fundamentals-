```python
class Person:
    def __init__(self, name, height, age, weight, gender):
        self.name = name
        self.height = height
        self.age = age
        self.weight = weight
        self.gender = gender

    def full_info(self):
        print(f'''Your name is {self.name}
Your age is {self.age}
Your height is {self.height}
Your weight is {self.weight}
Your gender is {self.gender}''')

    def set_age(self, new_age):
        if new_age < 0 or new_age > 120:
            print("Error: Please enter a realistic age between 0 and 120.")
        else:
            self.age = new_age
    def get_age(self):
        return self.age;
```

```python
p1 = Person("Shahzaib","168cm",19,78,"Male");
```

```python
p1.set_age(20);
```

### Inheritance 
-Parent class ki cheeze child class me ajai

```python
class Animal:
    def __init__(self, name, category, voice):
        self.name = name
        self.category = category
        self.voice = voice

    def speak(self):
        print(f"{self.name} is a {self.category} and it {self.voice}")

class animal(Animal):
    def intro(self):
        self.speak()

class animal2(Animal):
    def __init__(self, name, age, color, breed, gender):
        super().__init__(name, "Dog", "barks")
        self.age = age
        self.color = color
        self.breed = breed
        self.gender = gender

    def intro(self):
        self.speak()
        print(f"Age: {self.age}, Color: {self.color}, Breed: {self.breed}, Gender: {self.gender}")
```

```python
dog = animal("Saim","dog","barks")   
dog.intro()
```

```python
cat = animal("Mono","cat","Meows")
cat.intro()
```

```python
dog1 = animal2("Buddy", 3, "Golden", "Golden Retriever", "Male")
dog1.intro()
```

### POLYMORPHISIM 
Polymorphism ka matlab hai "one thing, many forms". OOP me ek method ya function same naam se hota hai, lekin different classes me different tareeqe se behave karta hai. Jaise `speak()` dog me "barks" kare aur cat me "meows" kare. Is se code zyada flexible aur reuseable hota hai.

```python
class Bird(Animal):
    def __init__(self, name, color):
        super().__init__(name, "Bird", "chirps")
        self.color = color

    def intro(self):
        self.speak()
        print(f"Color: {self.color}")

def animal_info(creature):
    creature.intro()

dog = animal("Saim", "dog", "barks")
cat = animal("Mono", "cat", "Meows")
parrot = Bird("Polly", "Green")

for pet in [dog, cat, parrot]:
    animal_info(pet)
```
